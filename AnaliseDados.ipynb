*Exercício 1:*

```python

import numpy as vet
import matplotlib.pyplot as plt

produto = vet.zeros(4)

produto[0] = input("Entre a quantidade de TV em estoque: ")
produto[1] = input("Entre a quantidade de Fogão em estoque: ")
produto[2] = input("Entre a quantidade de Notebook em estoque: ")
produto[3] = input("Entre a quantidade de Monitor em estoque: ")

labels = 'TV', 'Fogão', 'Notebook', 'Monitor'
plt.pie(produto, labels=labels, autopct='%1.1f%%')
plt.axis('equal')  
plt.show()
```

*Exercício 2:*

```python
import numpy as vet
import matplotlib.pyplot as plt

nota = vet.zeros(5)

for i in range(5):
  nota[i] = input("Entre com a nota: ")

plt.plot(nota,'bD')
plt.xlabel('Aluno')
plt.ylabel('Nota')
plt.ylim(0,10) 
plt.xlim(-1,5) 
plt.show()
```

```python
import numpy as vet
import matplotlib.pyplot as plt

idade = vet.zeros(10)

for i in range(10):
  idade[i] = input("Entre com a idade: ")

num_bins = 2
plt.hist(idade, num_bins, facecolor='red')
plt.show()
```
